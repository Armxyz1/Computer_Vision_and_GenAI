```python
from PIL import Image
from transformers import pipeline

vqa_pipeline = pipeline("visual-question-answering", model="Salesforce/blip-vqa-base")

# image =  Image.open("/content/SMART101-release-v1/SMART101-Data/1/img/puzzle_1_e_1.png")
# question = "Which object is linked to the hat? Answer from A: Flower, B: Book, C: Disk, D: drink, E:Ball"
```

```python
def check_ans(ans, user_ans, k):
    if user_ans in ['A', 'B', 'C', 'D', 'E']:
        if ans == user_ans:
            return True
    else:
        temp = data[k][ord(ans)-65+3]
        if user_ans.lower() == temp.lower():
            return True
        else:
            return False
```

```python
import csv
```

```python
tot = 0
acc = 0
for i in range(1,102):
  with open('SMART101-release-v1/SMART101-Data/'+str(i)+'/puzzle_'+str(i)+'.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
  data = data[1:]
  for k in range(10):
    tot +=1
    print(tot)
    ques = data[k][1]
    img_path =  "/content/SMART101-release-v1/SMART101-Data/"+str(i)+"/img/"+data[k][2]
    prompt = ques[:-1] +": "+ data[k][3] + ", " + data[k][4] + ", " + data[k][5] + ", " + data[k][6] + ", or " + data[k][7] + " ?"
    ans = data[k][8]
    img = Image.open(img_path)
    user_ans = vqa_pipeline(img, prompt, top_k=1)[0]['answer']
    if check_ans(ans, user_ans, k):
      acc+=1
```
